```python
import matplotlib.pyplot as plt
```

```python
skills = ['General Programming','Python','Visualization','Graphic Design','Data Analytics']
rate = [8,8,7,4,8]
```

```python
plt.bar(skills, rate, color ='powderblue', width = 0.7)
plt.xlabel("<--------- Skills -------->")
plt.ylabel("<--------- Experience Level --------->")
plt.title("Experience ratings of Abhinav in various skills")
plt.xticks(rotation=90)
plt.show()
```

Bar charts are useful to show discrete values/measure, though they may not prove very useful to show changing values and trends, like items measured over continuous intervals of time with chaning values at every instant (eg. half life of radioactive materials). Bar charts are also fairly easy to read. Since we have to visualize absolute value of experience rating from a scale of 0-10 for given skills, bar graph seemed like the apt choice of visualization in this case.

With the help of the above bar chart, I aim to visualize my experience in various specified skills like general programming, python, vizualization, etc. These skills are represented on the x-axis, where I have rotated the labels by 90 degrees to avoid overlapping of these labels because of larger spelling sizes. The x-axis is also labelled to indicate the listing of skills. The y-axis indicates the level of experience, the scale of rating being from 0-10. Like x-axis, the y-axis is also labelled with the title "Experience Level". I have given the entire visualization a title to indicate the main purpose of this representation. Additionally, I have increased the widths of the bar to make thicker, and have changed its colour from the default color to "powderblue" because I like colors that are on the cooler side like Teal, Aqua, Cornflowerblue, etc.
